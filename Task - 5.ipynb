## Task 5 : Exploratory Data Analysis - Sports 


## Step 1 : Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2 : Reading the Data

```python
df=pd.read_csv('matches.csv')
print(df)
print('-------------------------------------')
print('Dimenssion:',df.shape)
print('-------------------------------------')
print('Size:',df.size)
```

## Step 3 : Plotting the Data for Most successful Team

```python
teams=df['winner'].value_counts()
print('Teams with total number of Wins:')
print(teams)
print()
print('Data in graphical format:')
teams.plot.bar()
plt.xlabel('Teams')
plt.ylabel('Number of wins')
plt.title('Most successful teams in IPL from 2008-2019')
plt.show()
```

## Step 4 : Plotting the Data for Most number of Player of the Match award

```python
mom=df['player_of_match'].value_counts().nlargest(20)
print('Players with number of Player of the Match award:')
print(mom)
print()
print('Data in graphical format:')
mom.plot.bar()
plt.xlabel('Players')
plt.ylabel('Number of "Player of the Match" award')
plt.title('Top 20 players with most number of "Player of the Match" award in IPL from 2008-2019')
plt.show()
```

## Step 5 : Reading the Data

```python
df1=pd.read_csv('deliveries.csv')
print(df1)
print('-------------------------------------')
print('Dimenssion:',df1.shape)
print('-------------------------------------')
print('Size:',df1.size)
```

## Step 6 : Plotting the Data of Players with Most Runs

```python
players=df1['batsman'].value_counts().nlargest(20)
print('Batsmen with Runs:')
print(players)
print()
print('Data in Graphical format:')
players.plot.bar()
plt.xlabel('Players')
plt.ylabel('Runs')
plt.title(' Top 20 batsmen with most runs in IPL from 2008-2019')
plt.show()
```

## Step 7 : Plotting the details of Most Successful Bowler

```python
bowlers=df1.groupby(['bowler','dismissal_kind']).size().nlargest(50)
print('Bowlers with number of Dismissals and Kind of Dismissals')
print(bowlers)
print()
print('Data in Graphical format:')
bowlers.plot.bar()
plt.xlabel('Bowlers with kind of dismissal')
plt.ylabel('Number of dismissals')
plt.title(' Top bowlers with most dismissals in IPL from 2008-2019')
plt.show()
```
